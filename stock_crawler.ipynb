```python
url = "https://cn.investing.com/instruments/HistoricalDataAjax"


st_date = input("輸入開始日期:YYYY/mm/dd")
end_date = input("輸入結束日期:YYYY/mm/dd")
interval_type = int(input("類型 1=每日,2=每週,3=單月"))

if interval_type == 1:
    interval_sec = "Daily" 
elif interval_type == 2:
    interval_sec = "Weekly"
else:
    interval_sec = "Monthly"

datas = {
    "curr_id": "6408",
    "smlID": "1159963",
    "header": "AAPL历史数据",
    "st_date": st_date,
    "end_date": end_date,
    "interval_sec": interval_sec,
    "sort_col": "date",
    "sort_ord": "DESC",
    "action": "historical_data",
}

user_agent = UserAgent().random
headers = {
    "user-agent": user_agent,
    "x-requested-with": "XMLHttpRequest"
    
}

res = requests.post(url,data=datas,headers=headers)


soup = BeautifulSoup(res.text)
tr_list = soup.find_all("tr")
data_list = []
_len = len(tr_list)
for i,row in enumerate(tr_list):
    if i != 0: 
        td_list = row.find_all("td")
        if len(td_list) == 7:
            data = {}
            data["date"] = td_list[0].text
            data["closing_price"] = td_list[1].text
            data["opening_price"] = td_list[2].text
            data["high"] = td_list[3].text
            data["low"] = td_list[4].text
            data["trading_volume"] = td_list[5].text
            data["quote_change"] = td_list[6].text
            data_list.append(data)
     
data_json = json.dumps(data_list)
print(data_json)
```
